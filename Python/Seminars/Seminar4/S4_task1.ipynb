```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./GB.SEM.laptops-prices.csv', encoding='cp1251')
```

```python
df.head(2)
```

```python
sns.histplot(df['Memory_Amount'])
```

```python
plt.figure(figsize=(6, 3))
plt.hist(df['Memory_Amount'], bins=40)
plt.xlabel('Memory Amount, Gb')
plt.ylabel('Quantity of noteboks, Units')
plt.title('Memory distribution')
#plt.grid()
```

```python
plt.figure(figsize=(6, 3))
plt.hist(df['Price_euros'], bins=50)
plt.xlabel('Price, euros')
plt.ylabel('Quantity of noteboks, Units')
plt.title('Price distribution')
#plt.grid()
```
