```python
dfWithoutSamsung = df[df['Company'] != 'Samsung']
```

```python
dfWithoutSamsung.columns
```

```python
# Готовим данные для графика
data = pd.crosstab(df['Company'], df['Cpu_Company']).drop(columns='Samsung').reset_index()
data.head(3)
```

```python
# Строим столбчатую многорядную диаграмму

plt.figure(figsize=(6, 5))

# определяем кол-во делений
n_ticks = np.arange(len(data['AMD']))

# определяем сдвиг
offset = 0.2

# определяем ширину столбцов
w = 0.5

# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks - offset, data['AMD'], width=w)
plt.bar(n_ticks + offset, data['Intel'], width=w)

plt.title('Distribution of Cpu Company among notebook Producers')
plt.xlabel('Producers')
plt.ylabel('Units')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data['Company'], rotation = 50);  # добавляем метки делений
```
